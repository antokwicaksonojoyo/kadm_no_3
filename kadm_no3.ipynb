```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
records = []
for i in range(0, 2054) :
 records.append([str(store_data.values[i,j]) for j in range(0,10)])
```

```python
association_rules = apriori(records, min_support=0.2, min_confidence=0.2, min_left=0.2, min_lenght=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print((association_results))
```

```python
results =[]
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    
    value0 = str(item[0])
    value1 = str(item[1])
    value2 = str(item[1])[:6]
    value3 = str(item[2][0][2])[:6]
    value4 = str(item[2][0][3])[:6]
    
    rows = (value0,value1,value2,value3,value4)
    results.append(rows)
    label = ['title1','title2','Support','confidence','lift',]
    store_suggestion = pd.DataFrame.from_records(results,columns=label)
    print(store_suggestion)
```
